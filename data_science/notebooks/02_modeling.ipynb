## Logistic Regression 
- Predict the probability of a customer buying a shirt or not. 

**Feature Engineering**:
- Binary column
- Select relevant features
- One-Hot encode categorical
- Fill missing values wih 0

```python
# Binary column in df1 for is_shirt_purchase
df1 = df
df1['is_shirt_purchase'] = (df1['Category'] == 'T-shirts').astype(int)

# Select relevant features
features = ['Item price', 'Brand', 'Size', 'City', 'Boosting fee','time_to_sell_days']
X = df1[features]

# One-hot endcoding for categorical columns
X = pd.get_dummies(X, drop_first=True)

# Fill missing values
X = X.fillna(0)
```

**Train-Test Split**

```python
# Identify target variable
Y = df['is_shirt_purchase']

# Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(
X, Y, test_size=0.2, random_state=42)
```

## Decision Tree
- Predict likelyhood of products from the e-commerce shop selling less than 30 days.

```python
# Import decision tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

**Target Variable**

```python
df1['fast_sale'] = (df['time_to_sell_days'] < 30).astype(int)
```

**Select Features and Encode Categorical**

```python
# Select features and target variable
features_dt = ['Item price', 'Brand', 'Size', 'City', 'Boosting fee','time_to_sell_days']
X_dt = df1[features_dt]
Y_dt =df1['fast_sale']

# Encode categorical 
X_dt = pd.get_dummies(X_dt, drop_first=True)

# Fill missing
X_dt = X_dt.fillna(0)
```

**Train-Test Split**

```python
X_traindt, X_testdt, Y_traindt, Y_testdt = train_test_split(
    X_dt, Y_dt, test_size=0.2, random_state=42)
```

**Fit Model**

```python
dt_model = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_model.fit(X_traindt, Y_traindt)
```

**Predict and Evaluate**

```python
Y_preddt = dt_model.predict(X_testdt)
print("Accuracy:", accuracy_score(Y_testdt, Y_preddt))
print("Confusion Matrix:\n", confusion_matrix(Y_testdt, Y_preddt))
print("Classification Report:\n", classification_report(Y_testdt, Y_preddt))
```
